```python
import pandas as pd

# jobs = pd.read_excel(r'C:\Users\nfiru\Documents\Programmieren\TechLabs\Jobs during Corona\jobsInfo.xlsx')
jobs = pd.read_csv(r'C:\Users\nfiru\Documents\Programmieren\TechLabs\Jobs during Corona\18-05_jobsInfo.csv', index_col=0)
jobs = pd.DataFrame(jobs)

jobs = jobs.astype({'crawl_date':'datetime64'})
jobs['data_date'] = jobs['crawl_date'] - pd.to_timedelta(jobs['adsAge'], unit='d')
jobs['data_date'] = jobs['data_date'].dt.normalize()
jobs.set_index('data_date', inplace=True)
jobs = jobs.sort_index()
jobs = jobs[jobs.index.dayofweek < 5] # remove weekends

# jobs['Unnamed: 0'] = jobs['Unnamed: 0']*(-1)
# jobs.rename(columns = {'Unnamed: 0':'Past Days Included'}, inplace=True)
# jobs.set_index('Unnamed: 0')
jobs
```

```python
jobs.columns.tolist()
```

### Categories

```python
jobs.rename(columns={'jobs':'Total Jobs'}, inplace=True)

jobTypes = jobs.loc[:, 'Total Jobs':'Auf Kommission'].columns.tolist()
print("Job Types:")
print(jobTypes)

topCities = jobs.loc[:, 'München':'Oranienburg'].columns.tolist()
print('\n'"Top Cities:")
print(topCities)

#languages = jobs.loc[:, 'Deutsch':'Nederlands'].columns.tolist()
#print('\n'"Languages:")
#print(languages)
```

## New Job Types df

```python
df_jobTypes = jobs.loc[:,jobTypes]
df_jobTypes['Total Jobs'] = jobs['Total Jobs']
df_jobTypes
```

```python
# Reverse dataframe
# df_jobTypes_daily = df_jobTypes[::-1]

# Get difference
df_jobTypes_daily = df_jobTypes.diff(periods=-1) # difference with following row instead previous // last row = last row of df_jobTypes
df_jobTypes_daily
```

```python
# For later
df_jobTypes_daily.Vollzeit.pct_change() * 100
```

## Graphs for Job Types

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Define plot
plt.figure(figsize=(10,6))
plt.title("Daily new full-time jobs posted on Indeed.com")

sns.barplot(x=df_jobTypes_daily.index, y="Vollzeit", data=df_jobTypes_daily, color="blue")

plt.ylabel("Number of jobs posted")
plt.xlabel("Date posted")
plt.xticks(plt.xticks()[0], df_jobTypes_daily.index.date, rotation=45)
```

```python
# Define plot
plt.figure(figsize=(10,6))
plt.title("Daily new part-time jobs posted on Indeed.com")

sns.barplot(x=df_jobTypes_daily.index, y="Teilzeit", data=df_jobTypes_daily, color="red")

plt.ylabel("Number of jobs posted")
plt.xlabel("Date posted")
plt.xticks(plt.xticks()[0], df_jobTypes_daily.index.date, rotation=45)
```

```python
import numpy as np

ind = np.arange(len(df_jobTypes_daily['Vollzeit']))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, df_jobTypes_daily['Vollzeit'], width, label='Vollzeit')
rects2 = ax.bar(ind + width/2, df_jobTypes_daily['Teilzeit'], width, label='Teilzeit')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Jobs')
ax.set_title('Number of Full-time and Part-time Jobs Posted')
#ax.set_xticks(ind)
ax.set_xticklabels(df_jobTypes_daily.index.date, rotation=45)
ax.legend()
```

## New df Cities

```python
df_cities = jobs.loc[:,topCities]
df_cities
```

```python
# Get difference
df_cities_daily = df_cities.diff(periods=-1)
df_cities_daily
```

## Jobs per City

```python
ind = np.arange(len(df_cities_daily['Berlin']))  # the x locations for the groups
width = 0.4  # the width of the bars

fig, ax = plt.subplots(figsize=(18,10))
rects1 = ax.bar(ind - 2*width/4, df_cities_daily['München'], width, label='München')
rects2 = ax.bar(ind - width/4, df_cities_daily['Berlin'], width, label='Berlin')
rects3 = ax.bar(ind + 2*width/4, df_cities_daily['Frankfurt am Main'], width, label='Frankfurt')
rects4 = ax.bar(ind + width/4, df_cities_daily['Hamburg'], width, label='Hamburg')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Jobs')
ax.set_title('Total Jobs Posted in Top Cities')
#ax.set_xticks(ind)
ax.set_xticklabels(df_jobTypes_daily.index.date, rotation=45)
ax.legend()
```
